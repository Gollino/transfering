```python
import pandas as pd

import os

from tqdm import tqdm

import plotly.graph_objects as go
```

```python
df = pd.DataFrame()

for file in tqdm(os.listdir('database')):
    
    date = file.replace('SAMET_','').replace('T00.dat','')
    df_tmp = pd.read_csv(f'database/{file}')
    df_tmp['date'] = date
    df = df.append(df_tmp)
    
df.rename(columns={'Unnamed: 0':'state'}, inplace=True)
df.reset_index(drop=True, inplace=True)

df.to_csv('temperatura.csv', index=False)
```

```python
df.head()
```

```python
df['date'] = pd.to_datetime(df['date'], format='%Y%m%d')
```

```python
dict_sin = {'AM' : [ 'N',    'N'],
            'RR' : [ 'N',    '-'],
            'AP' : [ 'N',    'N'],
            'PA' : [ 'N',    'N'],
            'TO' : [ 'N',    'N'],
            'RO' : [ 'N', 'SECO'],
            'AC' : [ 'N', 'SECO'],
            'MA' : ['NE',    'N'],
            'PI' : ['NE',   'NE'],
            'CE' : ['NE',   'NE'],
            'RN' : ['NE',   'NE'],
            'PE' : ['NE',   'NE'],
            'PB' : ['NE',   'NE'],
            'SE' : ['NE',   'NE'],
            'AL' : ['NE',   'NE'],
            'BA' : ['NE',   'NE'],
            'DF' : ['CO', 'SECO'],
            'MT' : ['CO', 'SECO'],
            'MS' : ['CO', 'SECO'],
            'GO' : ['CO', 'SECO'],
            'SP' : ['SE', 'SECO'],
            'RJ' : ['SE', 'SECO'],
            'ES' : ['SE', 'SECO'],
            'MG' : ['SE', 'SECO'],
            'PR' : [ 'S',    'S'],
            'RS' : [ 'S',    'S'],
            'SC' : [ 'S',    'S']}

df['region_geo'] = [dict_sin[i][0] for i in df['state']]
df['region_sin'] = [dict_sin[i][1] for i in df['state']]
```

```python
import plotly.express as px

for region in ['SECO','S','NE','N']:
    dfx = df[df['region_sin'] == region]
    
    fig = px.line(dfx, x='date', y='tmed', color='state')
    fig.update_layout(title=region)
    fig.show()
```
